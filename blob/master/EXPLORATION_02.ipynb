### (1) load_digits : 손글씨 데이터 분류

```python
# (1) 필요한 모듈 import
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# (2) 데이터 준비
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
# (3) 데이터 이해하기

digits_data = digits.data
digits_label = digits.target
print('Target Names : ',digits.target_names)
print(digits.DESCR)
```

```python
#(4) train, test 데이터 분리
X_train,X_test,y_train,y_test = train_test_split(digits_data,
                                                 digits_label,
                                                test_size = 0.2,
                                                random_state = 30)
```

```python
#(5) 다양한 모델로 학습시켜보기
#Decision Tree,Random Forest,SVM,SGD Classifier,Logistic Regression

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression


decision_tree = DecisionTreeClassifier(random_state=30)
decision_tree.fit(X_train, y_train)
y_pred_1= decision_tree.predict(X_test)

random_forest = RandomForestClassifier(random_state=30)
random_forest.fit(X_train, y_train)
y_pred_2 = random_forest.predict(X_test)

svm_model = svm.SVC()
svm_model.fit(X_train,y_train)
y_pred_3 = svm_model.predict(X_test)

sgd_model = SGDClassifier()
sgd_model.fit (X_train,y_train)
y_pred_4 = sgd_model.predict(X_test)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred_5= logistic_model.predict(X_test)

```

```python
#(6) 모델을 평가해 보기
from sklearn.metrics import accuracy_score

#Decision Tree
accuracy_1 = accuracy_score(y_test, y_pred_1)
print("Accuracy of Decision Tree : {:.2f}".format(accuracy_1))
#Random Forest
accuracy_2 = accuracy_score(y_test, y_pred_2)
print("Accuracy of Random Forest :  {:.2f}".format(accuracy_2))

#SVM
accuracy_3 = accuracy_score(y_test, y_pred_3)
print("Accuracy of SVM : {:.2f}".format(accuracy_3))
#SGD
accuracy_4 = accuracy_score(y_test, y_pred_4)
print("Accuracy of SGD :  {:.2f}".format(accuracy_4))

#Logistic Regression
accuracy_5 = accuracy_score(y_test, y_pred_5)
print("Accuracy of Logistic Regression : {:.2f}".format(accuracy_5))


```

#### 평가지표 선택의 이유
imbalaced data도 아니고 label 의 class 가 여러개인 특성을 가진 데이터입니다. 값을 정확히 예측하는지만을 확인하면 되기 때문에 정확도(Accuracy)를 사용하여 평가 하였습니다.

### (3) load_breast_cancer : 유방암 데이터 분류

```python
#(1) 필요한 모듈 import하기
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# (2) 데이터 준비
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
```

```python
# (3) 데이터 이해하기

breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target
print('Target Names : ',breast_cancer.target_names)
print(breast_cancer.DESCR)
```

```python
#(4) train, test 데이터 분리
X_train,X_test,y_train,y_test = train_test_split(breast_cancer_data,
                                                breast_cancer_label,
                                                test_size = 0.2,
                                                random_state = 32)
```

```python
#(5) 다양한 모델로 학습시켜보기
#Decision Tree,Random Forest,SVM,SGD Classifier,Logistic Regression

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression


decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred_1= decision_tree.predict(X_test)

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred_2 = random_forest.predict(X_test)

svm_model = svm.SVC()
svm_model.fit(X_train,y_train)
y_pred_3 = svm_model.predict(X_test)

sgd_model = SGDClassifier()
sgd_model.fit (X_train,y_train)
y_pred_4 = sgd_model.predict(X_test)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred_5= logistic_model.predict(X_test)
```

```python
#(6) 모델을 평가해 보기
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

#Decision Tree
accuracy_1 = accuracy_score(y_test, y_pred_1)
print("Accuracy of Decision Tree : {:.2f}".format(accuracy_1))
print(classification_report(y_test, y_pred_1))
#Random Forest
accuracy_2 = accuracy_score(y_test, y_pred_2)
print("Accuracy of Random Forest :  {:.2f}".format(accuracy_2))
print(classification_report(y_test, y_pred_2))
#SVM
accuracy_3 = accuracy_score(y_test, y_pred_3)
print("Accuracy of SVM : {:.2f}".format(accuracy_3))
print(classification_report(y_test, y_pred_3))
#SGD
accuracy_4 = accuracy_score(y_test, y_pred_4)
print("Accuracy of SGD :  {:.2f}".format(accuracy_4))
print(classification_report(y_test, y_pred_4))
#Logistic Regression
accuracy_5 = accuracy_score(y_test, y_pred_5)
print("Accuracy of Logistic Regression : {:.2f}".format(accuracy_5))
print(classification_report(y_test, y_pred_5))
```

#### 평가지표 선택의 이유
breast_cancer같은 경우에는 accuracy 뿐만 아니라 Recall 값도 확인하기위해 classification_report를  출력하였습니다. 
recall은 대상 물체들을 빠뜨리지 않고 얼마나 잘 잡아내는지를 나타냅니다. 암에 대해 음성을 양성으로 오진하는것은 비교적 괜찮지만 양성을 음성으로 오진하면 문제가 있다고 볼 수 있습니다.
